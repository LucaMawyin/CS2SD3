```python
%%writefile producer_consumer.go

package main

import (
    "fmt"
    "math/rand/v2"
    "time"
)

type Product struct{
    ID int
    Status string
}

func main(){
    productList := Produce(20)
    Consume(productList)
    updateProductStatus(productList)
    reportProductStatus(productList)

    println("Done")
}

func updateProductStatus(productList []*Product){
    for _, product := range productList{
        time.Sleep(
            time.Duration(rand.IntN(300)) * 
            time.Millisecond,
        )

        status := []string{"Pending","Consuming","Consumed"}[rand.IntN(3)]

        product.Status = status
        fmt.Printf("Updating order %d with status: %s\n", product.ID, product.Status)
    }
}

func Consume(productList []*Product) {
    for _, product := range productList{
        time.Sleep(
            time.Duration(rand.IntN(500)) * 
            time.Millisecond,
        )
        fmt.Printf("Consuming order %d\n", product.ID)
    }
    
}

func Produce(count int) []*Product {
    productList := make([]*Product, count)

    for i := 0; i < count; i++ {
        productList[i] = &Product{
            ID: i,
            Status: "Pending",
        }
    }

    return productList
}

func reportProductStatus(productList []*Product){
    for i := 0; i < 5; i++{
        time.Sleep(1 * time.Second)
        fmt.Println("\n--- Status Report ---")
        for _, product := range productList{
            fmt.Printf("Order %d: %s\n", product.ID, product.Status)
        }
        fmt.Println("---------------------\n")
    }
}
```

```python
%%writefile producer_consumer.go

package main

import (
    "fmt"
    "math/rand/v2"
    "time"
    "sync"
)

type Product struct{
    ID int
    Status string
    mu sync.Mutex
}

var (
    totalUpdates int
    updateMutex sync.Mutex
)

func main(){

    var wg sync.WaitGroup
    wg.Add(3)

    productList := produce(20)

    for i := 0; i < 3; i++{
        go func(){
            defer wg.Done()
            for _, product := range productList{
                updateProductStatus(product)
            }
        }()
    }

    wg.Wait()
    reportProductStatus(productList)

    fmt.Println("Done")
    fmt.Printf("Total Updates: %d\n", totalUpdates)

}

func updateProductStatus(product *Product){

    product.mu.Lock()

    time.Sleep(
        time.Duration(rand.IntN(300)) * 
        time.Millisecond,
    )

    status := []string{"Pending","Consuming","Consumed"}[rand.IntN(3)]

    product.Status = status
    fmt.Printf("Updating order %d with status: %s\n", product.ID, product.Status)

    product.mu.Unlock()

    updateMutex.Lock()
    defer updateMutex.Unlock()
    currentUpdates := totalUpdates
    time.Sleep(5 * time.Millisecond)
    totalUpdates = currentUpdates + 1 
}

func consume(productList []*Product) {
    for _, product := range productList{
        time.Sleep(
            time.Duration(rand.IntN(500)) * 
            time.Millisecond,
        )
        fmt.Printf("Consuming order %d\n", product.ID)
    }
    
}

func produce(count int) []*Product {
    productList := make([]*Product, count)

    for i := 0; i < count; i++ {
        productList[i] = &Product{
            ID: i,
            Status: "Pending",
        }
    }

    return productList
}

func reportProductStatus(productList []*Product){
    fmt.Println("\n--- Status Report ---")
    for _, product := range productList{
        fmt.Printf("Order %d: %s\n", product.ID, product.Status)
    }
    fmt.Println("---------------------\n")
}
```
