We are now implementing channels. This allows goroutines to communicate information to one another

```python
%%writefile producer_consumer.go

package main

import (
    "fmt"
    "math/rand/v2"
    "time"
    "sync"
)

type Product struct{
    ID int
    Status string
    mu sync.Mutex
}


func main(){

    var wg sync.WaitGroup
    wg.Add(2)

    prodChan := make(chan *Product)

    go func(){
        defer wg.Done()
        for _, product := range produce(20){
            prodChan <- product
        }

        close(prodChan)

        fmt.Println("Done generating products")
    }()

    productList := produce(20)

    go consume(prodChan, &wg)

    wg.Wait()

    fmt.Println("Done")
}



func consume(prodChan <-chan *Prod, wg *sync.WaitGroup) { // The arrow specifies whether function sends or receives from channel

    defer wg.Done()

    for prod := range prodChan{
        time.Sleep(
            time.Duration(rand.IntN(500)) * 
            time.Millisecond,
        )
        fmt.Printf("Consuming order %d\n", product.ID)
    } 
}

func produce(count int) []*Product {
    productList := make([]*Product, count)

    for i := 0; i < count; i++ {
        productList[i] = &Product{
            ID: i,
            Status: "Pending",
        }
    }

    return productList
}

func reportProductStatus(productList []*Product){
    fmt.Println("\n--- Status Report ---")
    for _, product := range productList{
        fmt.Printf("Order %d: %s\n", product.ID, product.Status)
    }
    fmt.Println("---------------------\n")
}
```
